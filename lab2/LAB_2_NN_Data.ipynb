```python
from sklearn import datasets
```

make_regression():

```python
n_rows = 100000
n_cols = 30-1
data = datasets.make_regression(
    n_samples=n_rows, 
    n_features=n_cols, 
    n_informative=int(n_cols/2), 
    n_targets = int(n_cols/n_cols),
    bias=150,
    effective_rank = None, #сингулярні вектори
    tail_strength = 0.7, #важливість жирного шумового хвоста
    noise = 0.9, #шум
    shuffle = True,
    coef = False,
    random_state=42)
```

Data:


```python
import pandas as pd
df = pd.DataFrame(data[0])
df = df.add_prefix('x') #adding prefix to column names
df['y'] = data[1]
df.head()
```

```python
df.to_csv('data.csv', sep=';', index=False)
```
