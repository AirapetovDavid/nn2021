#  Matplotlib функція виду y=ax+b

```python
b = randint(0, 10)
k = randint(0, 10)
x = np.array([np.linspace(0, 10, 50)], ndmin=2).reshape((50,1))
y = k*x+b

plt.title("Лінійна функція y = {}*x+{}".format(k,b))
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.plot(x, y)
```

```python
x_new = np.reshape(x, (-1, 2))
y_new = np.reshape(y, (-1, 2))
```

```python
reg = LinearRegression()
reg.fit(x_new ,y_new)
Y_hat = reg.predict(x_new)
print("Y початкове:", y)
print("-------------------------------------------")
print("Y, що обраховано лінійною регресією:", Y_hat)
```

```python
square_err = mean_squared_error(y_new,Y_hat)
print("MSE = ", square_err )
abs_err = mean_absolute_error(y_new,Y_hat)
print("MAE = ", abs_err )
```

```python
plt.scatter(x, y,  color='black')
plt.grid()
plt.plot(x_new, Y_hat, color='blue', linewidth=3)
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
plt.show()
```

```python
headers = ['X', 'Y', 'Y_hat', 'MSE', 'MAE']
Y_hat = Y_hat.flatten()
with open('matplotlab.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(headers)
    for i in range(25):
        data = [x[i],y[i],round(Y_hat[i],3),square_err,abs_err]
        writer.writerow(data)
```

# Classification

```python
data = pd.read_csv("mobile.csv")
data.head()
```

```python
data['price_range'].value_counts()
```

```python
X = data.drop(['price_range'], axis=1)
Y = data['price_range']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X , Y,test_size=0.25,random_state=52)
```

```python
norm = preprocessing.MinMaxScaler()
norm.fit(X_train)
X_train_norm = norm.transform(X_train)
```

```python
norm.fit(X_test)
X_test_norm = norm.transform(X_test)
```

```python
model = svm.SVC()
model.fit(X_train_norm,Y_train)
Y_hat = model.predict(X_test_norm)
print("Accuracy:")
print((metrics.accuracy_score(Y_test, Y_hat))*100,"%")
```

```python
conf_matrix = metrics.confusion_matrix(Y_test,Y_hat)
conf_matrix = pd.DataFrame(conf_matrix, index= [0,1,2,3], columns=[0,1,2,3])
conf_matrix
```
